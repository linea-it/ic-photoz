```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("ic-photoz/Fase4-Artigo/sample_dp02.csv")
print("quantidade de itens", len(df))
df.head()
```

```python
plt.figure(figsize=[8,8])
plt.suptitle("Sample in DP 0.2", fontsize=18)

plt.plot(df.coord_ra, df.coord_dec, f".", color="midnightblue", alpha=0.2)
plt.xlabel("RA", fontsize=18)
plt.ylabel("DEC", fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.xticks(fontsize=16)
plt.tight_layout()
```

```python
def qtd_cor(key1, key2, x_min=-5, x_max=5, y_min=0, y_max=500, x_size=16):
    plt.hist(df[f"mag_{key1}"]-df[f"mag_{key2}"],  bins="fd", density=True)

    plt.xlabel(f"color ({key1}-{key2})", fontsize=16)

    plt.xticks(fontsize=x_size)
    plt.yticks(fontsize=16)
    
    #plt.xlim(x_min, x_max)
    #plt.ylim(y_min, y_max)
    #plt.vlines(0, -10, 10)
    #plt.hlines(0, -10, 10)
```

```python
plt.figure(figsize=[8,8], dpi=300)
plt.suptitle("Quantidade x Cor", fontsize=16)

plt.subplot(2,2,1)
qtd_cor("u", "g")

plt.subplot(2,2,2)
qtd_cor("g", "r")

plt.subplot(2,2,3)
qtd_cor("r", "i")

plt.subplot(2,2,4)
qtd_cor("i", "z")

plt.tight_layout()
```
